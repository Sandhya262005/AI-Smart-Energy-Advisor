```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Units_Consumed": [210, 230, 250, 270, 300, 320]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure()
plt.plot(df["Month"], df["Units_Consumed"])
plt.xlabel("Month")
plt.ylabel("Electricity Units Consumed")
plt.title("Monthly Energy Consumption Pattern")
plt.show()
```

```python
average_usage = df["Units_Consumed"].mean()

if average_usage > 250:
    recommendation = """
    High energy consumption detected.
    AI Recommendations:
    - Use LED lighting
    - Avoid peak-hour appliance usage
    - Consider rooftop solar panels
    - Use energy-efficient appliances
    """
else:
    recommendation = """
    Energy consumption is optimal.
    Continue using energy-efficient practices.
    """

print(recommendation)
```

```python
def energy_advisor(units):
    if units < 150:
        return """
        Energy Usage Level: Low
        Suggestions:
        - Continue energy-efficient practices
        - Consider solar lighting for outdoor use
        """
    elif units < 300:
        return """
        Energy Usage Level: Moderate
        Suggestions:
        - Use energy-efficient appliances
        - Avoid peak-hour usage
        - Switch to LED lighting
        """
    else:
        return """
        Energy Usage Level: High
        Suggestions:
        - High consumption detected
        - Use energy-efficient appliances
        - Shift heavy usage to non-peak hours
        - Consider rooftop solar panels
        """

# USER INPUT
user_units = int(input("Enter your monthly electricity consumption (units): "))
print(energy_advisor(user_units))
```
